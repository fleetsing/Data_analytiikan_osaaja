```python
import sqlite3
```

```python
connection = sqlite3.connect("test.db")
```

```python
c = connection.cursor()
```

```python
c.execute('''CREATE TABLE TEST
             ([generated_id] INTEGER PRIMARY KEY, [Person] text, [Department] Text, [Title] text, [Salary] integer)''')
```

```python
connection.commit()
```

```python
read_hr = pd.read_csv("hr.csv")
read_hr.to_sql('TEST', connection, if_exists='append', index = False)
```
