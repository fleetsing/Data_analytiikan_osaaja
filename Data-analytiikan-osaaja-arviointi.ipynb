Tässä harjoituksessa kiinnitämme huomiota yleistykseen ja datan jakoon opetusjoukkoon ja testijoukkoon. Esimerkkinä käytämme aikaisempaa luokitteluongelmaa.

Ensin tarvitaan kuitenkin tavalliset kirjastot, josta saamme tietorakenteita ja funktiota käyttöön:

```python
import numpy as np
import pandas as pd
```

```python
# Esimerkkidatat täältä: 
from sklearn import datasets, neighbors

# Regressiomallien arviontia (ei tänään):
from sklearn.metrics import mean_squared_error, r2_score

# Jako opetus- ja testijoukkoihin:
from sklearn.model_selection import train_test_split
```

Otetaan käyttöön esimerkiksi soveltuva irisdata

```python
# Ladataan iris -datajoukko:
from sklearn.datasets import load_iris

(iris_X, iris_luokka) = load_iris(return_X_y=True)
```

Tehtävä 1. Jaa data-aineisto kahteen osaan, opetusjoukkoon ja testijoukkoon. Käytä siihen funktiota train_test_split, dokumentaatio löytyy täältä: https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

```python
iris_X_opetus1, iris_X_testi1, iris_luokka_opetus1, iris_luokka_testi1 = train_test_split(
    iris_X, iris_luokka, test_size = 0.33, random_state=42)

# Nyt pitäisi olla seuraavat muuttujat käytössä:
# iris_X_opetus
# iris_X_testi
# iris_luokka_opetus
# iris_luokka_testi
```

```python
# Luodaan ensin luokittelija:
luokittelija1 = neighbors.KNeighborsClassifier(n_neighbors=3)

# Opetetaan luokittelija data-aineistolla:
luokittelija1.fit(iris_X_opetus1, iris_luokka_opetus1)

# Luokitellaan testi_aineistolla:
opetus_luok1 = luokittelija1.predict(iris_X_opetus1)
testi_luok1 = luokittelija1.predict(iris_X_testi1)
```

Tehtävä 2. Seuraavaksi tulisi arvioida tuloksia. Laske opetusjoukon oikeinluokittelu absoluuttisina määrinä sekä osuutena (katso edellisen harjoituksen notebookista). Seuraavaksi laske testijoukon oikeinluokittelu absoluuttisina määrinä sekä osuutena (katso edellisen harjoituksen notebookista).

```python
# Ratkaisu tähän:

opetus_abs1 = np.sum(opetus_luok1 == iris_luokka_opetus1)
print(opetus_abs1)

opetus_suhd1 = np.sum(opetus_luok1 == iris_luokka_opetus1)/len(iris_luokka_opetus1)
print(opetus_suhd1)

testi_abs1 = np.sum(testi_luok1 == iris_luokka_testi1)
print(testi_abs1)

testi_suhd1 = np.sum(testi_luok1 == iris_luokka_testi1)/len(iris_luokka_testi1)
print(testi_suhd1)
```
