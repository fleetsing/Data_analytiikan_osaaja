## Tarvittavat kirjastot

```python
# Mieti mitä kirjastoja tarvitset luokitteluongelmassasi
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression, Lars
from sklearn.svm import SVR
```

## Lue data

```python
# Lataa data työtilaan joko tekstitiedostona, tietokannasta tai kirjastosta

boston_x, boston_y = datasets.load_boston(return_X_y = True)
```

## Tarvittava esikäsittely

```python
# Esikäsittelyä, esim. datan normalisointia, muuttujien poimintaa taulukosta
# Ennustusongelmissa esim. x,y oikeisiin muuttujiin

print(boston_x.shape)
print(boston_y.shape)
```

# lm

```python
lm_opetus_mse = np.zeros(10)
lm_validointi_mse = np.zeros(10)

lm_opetus_r2 = np.zeros(10)
lm_validointi_r2 = np.zeros(10)

for i in range(1,10):
    data_x_opetus, data_x_validointi, data_y_opetus, data_y_validointi = train_test_split(
            boston_x, boston_y, test_size=0.2, random_state=42*i)
    
    lm = LinearRegression(fit_intercept=True)
    lm.fit(data_x_opetus, data_y_opetus)
    
    ennustukset_opetus = lm.predict(data_x_opetus)
    lm_opetus_mse[i] = mean_squared_error(data_y_opetus, ennustukset_opetus)
    lm_opetus_r2[i] = r2_score(data_y_opetus, ennustukset_opetus)
    
    ennustukset_validointi = lm.predict(data_x_validointi)
    lm_validointi_mse[i] = mean_squared_error(data_y_validointi, ennustukset_validointi)
    lm_validointi_r2[i] = r2_score(data_y_validointi, ennustukset_validointi)
```

# svr_linear

```python
svr_linear_opetus_mse = np.zeros(10)
svr_linear_validointi_mse = np.zeros(10)

svr_linear_opetus_r2 = np.zeros(10)
svr_linear_validointi_r2 = np.zeros(10)

for i in range(1,10):
    data_x_opetus, data_x_validointi, data_y_opetus, data_y_validointi = train_test_split(
            boston_x, boston_y, test_size=0.2, random_state=42*i)
    
    svr_linear = SVR(kernel='linear', C=100, gamma=0.1, epsilon=0.1)
    svr_linear.fit(data_x_opetus, data_y_opetus)
    
    ennustukset_opetus = svr_linear.predict(data_x_opetus)
    svr_linear_opetus_mse[i] = mean_squared_error(data_y_opetus, ennustukset_opetus)
    svr_linear_opetus_r2[i] = r2_score(data_y_opetus, ennustukset_opetus)
    
    ennustukset_validointi = svr_linear.predict(data_x_validointi)
    svr_linear_validointi_mse[i] = mean_squared_error(data_y_validointi, ennustukset_validointi)
    svr_linear_validointi_r2[i] = r2_score(data_y_validointi, ennustukset_validointi)
```

# svr_poly

```python
svr_poly_opetus_mse = np.zeros(10)
svr_poly_validointi_mse = np.zeros(10)

svr_poly_opetus_r2 = np.zeros(10)
svr_poly_validointi_r2 = np.zeros(10)

for i in range(1,10):
    data_x_opetus, data_x_validointi, data_y_opetus, data_y_validointi = train_test_split(
            boston_x, boston_y, test_size=0.2, random_state=42*i)
    
    svr_poly = SVR(kernel='poly', C=100, gamma=0.1, epsilon=0.1)
    svr_poly.fit(data_x_opetus, data_y_opetus)
    
    ennustukset_opetus = svr_poly.predict(data_x_opetus)
    svr_poly_opetus_mse[i] = mean_squared_error(data_y_opetus, ennustukset_opetus)
    svr_poly_opetus_r2[i] = r2_score(data_y_opetus, ennustukset_opetus)
    
    ennustukset_validointi = svr_poly.predict(data_x_validointi)
    svr_poly_validointi_mse[i] = mean_squared_error(data_y_validointi, ennustukset_validointi)
    svr_poly_validointi_r2[i] = r2_score(data_y_validointi, ennustukset_validointi)
```

# svr_rbf

```python
svr_rbf_opetus_mse = np.zeros(10)
svr_rbf_validointi_mse = np.zeros(10)

svr_rbf_opetus_r2 = np.zeros(10)
svr_rbf_validointi_r2 = np.zeros(10)

for i in range(1,10):
    data_x_opetus, data_x_validointi, data_y_opetus, data_y_validointi = train_test_split(
            boston_x, boston_y, test_size=0.2, random_state=42*i)
    
    svr_rbf = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=0.1)
    svr_rbf.fit(data_x_opetus, data_y_opetus)
    
    ennustukset_opetus = svr_rbf.predict(data_x_opetus)
    svr_rbf_opetus_mse[i] = mean_squared_error(data_y_opetus, ennustukset_opetus)
    svr_rbf_opetus_r2[i] = r2_score(data_y_opetus, ennustukset_opetus)
    
    ennustukset_validointi = svr_rbf.predict(data_x_validointi)
    svr_rbf_validointi_mse[i] = mean_squared_error(data_y_validointi, ennustukset_validointi)
    svr_rbf_validointi_r2[i] = r2_score(data_y_validointi, ennustukset_validointi)
```

# svr_sigmoid

```python
svr_sigmoid_opetus_mse = np.zeros(10)
svr_sigmoid_validointi_mse = np.zeros(10)

svr_sigmoid_opetus_r2 = np.zeros(10)
svr_sigmoid_validointi_r2 = np.zeros(10)

for i in range(1,10):
    data_x_opetus, data_x_validointi, data_y_opetus, data_y_validointi = train_test_split(
            boston_x, boston_y, test_size=0.2, random_state=42*i)
    
    svr_sigmoid = SVR(kernel='sigmoid', C=100, gamma=0.1, epsilon=0.1)
    svr_sigmoid.fit(data_x_opetus, data_y_opetus)
    
    ennustukset_opetus = svr_sigmoid.predict(data_x_opetus)
    svr_sigmoid_opetus_mse[i] = mean_squared_error(data_y_opetus, ennustukset_opetus)
    svr_sigmoid_opetus_r2[i] = r2_score(data_y_opetus, ennustukset_opetus)
    
    ennustukset_validointi = svr_sigmoid.predict(data_x_validointi)
    svr_sigmoid_validointi_mse[i] = mean_squared_error(data_y_validointi, ennustukset_validointi)
    svr_sigmoid_validointi_r2[i] = r2_score(data_y_validointi, ennustukset_validointi)
```
